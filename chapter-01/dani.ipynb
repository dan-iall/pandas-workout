```python
import pandas as pd
import numpy as np
```

```python
g = np.random.default_rng(0)
g
```

```python
scores = g.integers( 40 , 60 , 10 )
print (scores)
```

```python
months = 'Sep Oct Nov Dec Jan Feb Mar Apr May Jun'.split()
print(months)
```

```python
s = pd.Series(scores,index = months)
print(s)
```

```python
year_mean = s.mean()
print(year_mean)
```

```python
to_80 =  80 - year_mean
```

```python
s = s + to_80
s
```

```python
year_mean = s.mean()
print(year_mean)
```
