```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
filename = '../data/celebrity_deaths_2016.csv'

df = pd.read_csv(filename,
                usecols=['dateofdeath', 'age'])

df.head()
```

```python
# Make the "month" column the index of the data frame
df['month'] = df['dateofdeath'].str.slice(5,7)
df = df.set_index('month')
```

```python
# Sort the data frame by the index
df = df.sort_index()
```

### ðŸ•µï¸â€â™€ï¸ Bellonda's Logic Decoder
**The Business Goal:** To sanitize the 'age' column so we can perform mathematical operations, ensuring that any non-numeric garbage data is safely neutralized rather than crashing our code.

**The Syntax Anatomy:**
* `pd.to_numeric(...)` â†’ **The Converter**: The specialized function that attempts to turn text into numbers.
* `df['age']` â†’ **The Raw Input**: The column containing potentially mixed data types (strings and numbers).
* `errors='coerce'` â†’ **The Safety Valve**: The setting that says "If a value is invalid, force it to be `NaN` (Not a Number) instead of raising an error."

> **ðŸ’¡ The Wisdom Check:**
> "`errors='coerce'` is the data scientist's version of a quality filter. We are accepting that we will lose some data (the garbage) in exchange for a column we can actually trust and calculate with."

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```
