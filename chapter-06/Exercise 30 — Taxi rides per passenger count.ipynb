```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
filename = '../data/nyc_taxi_2019-01.csv'

df = pd.read_csv(filename, usecols=['passenger_count', 'trip_distance', 'total_amount'])

df.head()
```

### ðŸ•µï¸â€â™€ï¸ Bellonda's Logic Decoder
**The Business Goal:** To profile our revenue based on customer group size. This tells us the average ticket value for a solo rider versus a full car.

**The Syntax Anatomy:**
* `.groupby('passenger_count')` â†’ **The Segmentation**: We split our data into distinct buckets based on group size.
* `['total_amount']` â†’ **The Metric**: We isolate the financial column to perform calculations on.
* `.mean()` â†’ **The Aggregator**: We summarize the many values in each bucket into a single representative average.

> **ðŸ’¡ The Wisdom Check:**
> "By chaining these three, we turn millions of rows into a simple report: '1 Passenger = $X, 2 Passengers = $Y'. This is the heartbeat of data analysis."

```python
# For each number of passengers, find the mean cost of a taxi ride.
# Sort this result from lowest (i.e., cheapest) to highest (i.e., most expensive).

df.groupby('passenger_count')['total_amount'].mean()
```

```python
# Sort the results once again, in increasing number of passengers.

df.groupby('passenger_count')['total_amount'].mean().sort_index()
```

### ðŸ•µï¸â€â™€ï¸ Bellonda's Logic Decoder
**The Business Goal:** To determine if "Long Haul" trips attract different group sizes than "Short Haul" trips. We accept that raw distance is too noisy, so we categorize trips into three distinct "products" (Short, Medium, Long) to analyze passenger density per product type.

**The Syntax Anatomy:**
* `pd.cut(...)` â†’ **The Segmenter** (Transforms continuous numbers into categorical buckets)
* `bins=[..., 2, 10, ...]` â†’ **The Thresholds** (The specific mile-markers where a "Short" trip becomes a "Medium" trip)
* `.groupby(...).mean()` â†’ **The Pivot** (Calculates the average metric for each segment)
* `.sort_values(ascending=False)` â†’ **The Ranking** (Immediately visualizes which segment has the highest passenger density)

> **ðŸ’¡ The Wisdom Check:**
> When using `pd.cut`, always double-check your bin edges. Here we use the `.min()` and `.max()` of the data to ensure NO rows are dropped (returned as NaN) because they fell outside your manual range (e.g., a 0.1-mile trip or a 100-mile trip).

```python
# Now create a new column, `trip_length`, in which the values will be `short`
# (< 2 miles), `medium` (> 2 miles and <= 10 miles), or `long` (> 10 miles). 
# What was the average number of passengers per trip length category? Sort 
# this result from highest (greatest number of passengers) to lowest
# (smallest number of passengers).

df['trip_distance_group'] = pd.cut(df['trip_distance'], 
                                   [df['trip_distance'].min(), 2, 10, 
                                    df['trip_distance'].max()],
                                  labels=['short', 'medium', 'long'])
df.groupby('trip_distance_group')['passenger_count'].mean().sort_values(ascending=False)
```
